```python
#Dependencies
import pandas as pd 
```

```python
#get the csv files
csvpath1 = "C:/Users/chnze/pandas-challenge/Instructions/PyCitySchools/Resources/schools_complete.csv"
csvpath2 = "C:/Users/chnze/pandas-challenge/Instructions/PyCitySchools/Resources/students_complete.csv"
schools_df = pd.read_csv(csvpath1,encoding="utf-8")
students_df = pd.read_csv(csvpath2,encoding="utf-8")
```

```python
#merge the two csv files
ss_df = pd.merge(schools_df,students_df,on="school_name")
ss_df
ss_df.sort_values(by=['school_name'])
```

```python
#District Summary

#Total schools
total_schools = len(ss_df["school_name"].unique())


#Total students
total_students = len(ss_df["student_name"])

#Total budget
total_budget =ss_df["budget"].unique()
total_budget1 = total_budget.sum()

#Average math score
average_math = ss_df["math_score"].mean()

#Average reading school
average_reading = ss_df["reading_score"].mean()

#%passing math(the percentage of students who passed math)
greater_math70 = ss_df.loc[ss_df["math_score"] >=70,:]
pass_math = len(greater_math70)/total_students,


#passing reading(the percentage of students who passed math)
greater_reading70 = ss_df.loc[ss_df["reading_score"]>=70,:]
pass_reading = len(greater_reading70)/total_students,


#overall passing(the percentage of students who passed math AND reading)
math_reading_pass = ss_df.loc[(ss_df["reading_score"]>= 70) & (ss_df["math_score"]>= 70),:]
len(math_reading_pass)

overall_passing = len(math_reading_pass)/total_students

```

```python
#DataFrame
District_Summary = pd.DataFrame({"Total schools":[total_schools],"Total students": total_students,"Total budget":total_budget1,
                                "Average math score": average_math,"Average reading score":average_reading, "Passing math%": pass_math,
                                "Passing reading%":pass_reading,"Passing overall%":overall_passing})

District_Summary["Passing math%"] = (District_Summary["Passing math%"]*100).map("{:.2f}%".format)
District_Summary["Passing reading%"] = (District_Summary["Passing reading%"]*100).map("{:.2f}%".format)
District_Summary["Passing overall%"] = (District_Summary["Passing overall%"]*100).map("{:.2f}%".format)
District_Summary["Total budget"] = District_Summary["Total budget"].map("${:,.2f}".format)
District_Summary["Total students"] = District_Summary["Total students"].map("{:,}".format)
District_Summary
```

```python
#School Summary

bb = schools_df
bb1 = bb[['school_name','type','size','budget']]
bb2 = bb1.sort_values('school_name',ascending=True)
bb3 = list(bb2['budget'])



#Per student budget
budget_groupby_df = ss_df.groupby(['school_name'])
budget_groupby_df2 = budget_groupby_df.mean()
budget_groupby_df2 
per_student_budget = budget_groupby_df2['budget']/budget_groupby_df2['size']


#average math score
student_groupby_1 = ss_df.groupby(['school_name'])
student_groupby_2 = student_groupby_1.mean()
#student_groupby_3 = student_groupby_2["math_score"]
sg3m = list(student_groupby_2['math_score'])

#average reading score
sg3r = list(student_groupby_2['reading_score'])

#%passing math
pass_math = ss_df.loc[(ss_df["math_score"]>=70),:]
pass_math1 = pass_math.groupby(['school_name'])
pass_math2 = pass_math1.count()
pass_math22 = pass_math1.mean()
pass_math_total = pass_math2['math_score']/pass_math22['size']
pmt = list(pass_math_total)

#%passing reading
pass_reading = ss_df.loc[(ss_df["reading_score"]>=70),:]
pass_read1 = pass_reading.groupby(['school_name'])
pass_read2 = pass_read1.count()
pass_read22 = pass_read1.mean()
pass_read_total = pass_read2['math_score']/pass_read22['size']
prt = list(pass_read_total)

#%overallpassing
pass_overall = ss_df.loc[((ss_df["reading_score"]>=70) & (ss_df["math_score"]>=70)),:]
pass_overall1 = pass_overall.groupby(['school_name'])
pass_overall2 = pass_overall1.count()
pass_overall22 = pass_overall1.mean()
pass_overall_total = pass_overall2['math_score']/pass_overall22['size']
pot = list(pass_overall_total)

sort_1 = schools_df.sort_values(by=['school_name'])




```

```python
numberdata = pd.DataFrame({"budget":bb3,"Per student budget":per_student_budget,"Average math score":sg3m,"Average reading score":sg3r,
                           "Passing math%":pmt, "Passing reading%":prt,"Passing overall%":pot})

numberdata


School_Summary = pd.merge(bb1,numberdata,on="budget")
School_Summary["Passing math%"] = (School_Summary["Passing math%"]*100)
School_Summary["Passing reading%"] = (School_Summary["Passing reading%"]*100)
School_Summary["Passing overall%"] = (School_Summary["Passing overall%"]*100)
School_Summary["Per student budget"] = School_Summary ["Per student budget"].map("${:.2f}".format)
School_Summary["budget"] = School_Summary ["budget"].map("${:,.2f}".format)



School_Summary.set_index('school_name',inplace=True)
School_Summary.sort_values(by="school_name",ascending=True)
School_Summary.rename(columns=({'school_name': "School Name",'type':'School type','size':"Total Student",
                                'budget':'Total Budget'}))
```
